### 2. A partir du texte de la colonne cast, créer deux nouvelles colonnes dans le tableau credits qui contiennent respectivement le nom du personnage principal, et de l’acteur ou l’actrice qui l’incarne.

```python
import json

def get_main_cast_info(cast_str):
    # Vérifie si la chaîne de caractères est vide, NaN ou non une chaîne
    if pd.isna(cast_str) or not isinstance(cast_str, str):
        return "NA", "NA"
    try:
        cast_list = json.loads(cast_str)
        if cast_list:
            # Recherche du membre du casting avec l'ordre 0 (personnage/acteur principal)
            main_cast = next((item for item in cast_list if item.get('order') == 0), None)
            if main_cast:
                return main_cast.get('character', "NA"), main_cast.get('name', "NA")
            else:
                # Si aucun ordre 0, prend le premier membre du casting comme principal
                first_cast = cast_list[0]
                return first_cast.get('character', "NA"), first_cast.get('name', "NA")
        else:
            return "NA", "NA"
    except (json.JSONDecodeError, TypeError):
        # Gère les cas où cast_str n'est pas un JSON valide ou d'autres erreurs
        return "NA", "NA"

# Applique la fonction pour créer les nouvelles colonnes
credits[['main_character', 'main_actor']] = credits['cast'].apply(lambda x: pd.Series(get_main_cast_info(x)))

print("De nouvelles colonnes 'main_character' et 'main_actor' ont été ajoutées au DataFrame 'credits' :")
print(credits[['title', 'main_character', 'main_actor']].head(10))
```

### 3. Extraction de la colonne crew du tableau credits l’information du nom du réalisateur ou de la réalisatrice du film, et création d'une colonne pour cette information dans le fichier.

```python
def get_director_name(crew_str):
    # Vérifie si la chaîne de caractères est vide, NaN ou non une chaîne
    if pd.isna(crew_str) or not isinstance(crew_str, str):
        return "NA"
    try:
        crew_list = json.loads(crew_str)
        if crew_list:
            # Recherche du membre de l'équipe dont le job est 'Director'
            director = next((item for item in crew_list if item.get('job') == 'Director'), None)
            if director:
                return director.get('name', "NA")
        return "NA"
    except (json.JSONDecodeError, TypeError):
        # Gère les cas où crew_str n'est pas un JSON valide ou d'autres erreurs
        return "NA"

# Applique la fonction pour créer la nouvelle colonne 'director'
credits['director'] = credits['crew'].apply(lambda x: get_director_name(x))

print("Une nouvelle colonne 'director' a été ajoutée au DataFrame 'credits' :")
print(credits[['title', 'director']].head(10))
```

### 4. Extraction de genre (action, aventure, comédie, etc) du film dans une nouvelle colonne ’genre’ du tableau movies

```python
def get_genres(genres_str):
    # Vérifie si la chaîne de caractères est vide, NaN ou non une chaîne
    if pd.isna(genres_str) or not isinstance(genres_str, str):
        return "NA"
    try:
        genres_list = json.loads(genres_str)
        if genres_list:
            # Extrait les noms des genres et les joint avec une virgule
            return ", ".join([genre.get('name', 'NA') for genre in genres_list])
        else:
            return "NA"
    except (json.JSONDecodeError, TypeError):
        # Gère les cas où genres_str n'est pas un JSON valide ou d'autres erreurs
        return "NA"

# Applique la fonction pour créer la nouvelle colonne 'genre'
movies['genre'] = movies['genres'].apply(lambda x: get_genres(x))

print("Une nouvelle colonne 'genre' a été ajoutée au DataFrame 'movies' :")
print(movies[['title', 'genre']].head())
```

#### e- Compare la réponse avec le nom du film et, selon le cas, affiche la sortie Gagné ! ou la sortie Perdu, et donne une ou plusieurs nouvelles informations sur le film pour aider le joueur et le laisser deviner à nouveau.

```python
def play_5():
    # Dictionnaire des fournisseurs d'indices pour formater l'affichage
    hint_providers = {
        'first_letter_title': lambda movie_info: f"Première lettre du titre : {movie_info['title'][0].upper()}",
        'budget': lambda movie_info: f"Budget : {movie_info['budget'] / 1_000_000:.1f} millions USD" if movie_info['budget'] > 0 else "Budget : Non spécifié",
        'genre': lambda movie_info: f"Genre : {movie_info['genre']}",
        'director': lambda movie_info: f"Réalisateur : {movie_info['director']}",
        'main_actor': lambda movie_info: f"Acteur principal : {movie_info['main_actor']}",
        'release_date': lambda movie_info: f"Date de sortie : {movie_info['release_date']}",
        'overview': lambda movie_info: f"Synopsis (partiel) : {movie_info['overview'][:100]}..." if pd.notna(movie_info['overview']) and len(str(movie_info['overview'])) > 100 else f"Synopsis : {movie_info['overview']}",
        'popularity': lambda movie_info: f"Popularité : {movie_info['popularity']:.2f}",
        'vote_average': lambda movie_info: f"Note moyenne : {movie_info['vote_average']:.1f}",
        'original_language': lambda movie_info: f"Langue originale : {movie_info['original_language'].upper()}"
    }

    # Séquence d'indices pour chaque difficulté (ordre de révélation)
    difficulty_hint_sequence = {
        'difficile': [
            'first_letter_title', 'budget',
            'genre', 'director', 'main_actor', 'release_date', 'overview', 'popularity', 'vote_average', 'original_language'
        ],
        'moyen': [
            'first_letter_title', 'genre', 'director', 'budget',
            'main_actor', 'release_date', 'overview', 'popularity', 'vote_average', 'original_language'
        ],
        'facile': [
            'first_letter_title', 'genre', 'budget', 'director', 'main_actor',
            'release_date', 'overview', 'popularity', 'vote_average', 'original_language'
        ]
    }

    # Nombre d'indices affichés initialement pour chaque difficulté
    initial_hints_count = {
        'difficile': 2,
        'moyen': 3,
        'facile': 4
    }

    # Nombre maximal d'essais pour chaque difficulté
    max_attempts_dict = {
        'difficile': 3, # Par exemple, 3 essais pour difficile
        'moyen': 5,    # 5 essais pour moyen
        'facile': 7     # 7 essais pour facile
    }

    # --- Partie 1: Sélection de la difficulté ---
    difficulty = ""
    while True:
        print("Choisissez un niveau de difficulté : difficile, moyen ou facile.")
        choice = input().lower().strip()
        if choice in ['difficile', 'moyen', 'facile']:
            difficulty = choice
            print(f"Vous avez choisi le mode : {difficulty}")
            break
        else:
            print("Choix invalide. Veuillez réessayer.")

    # --- Partie 2: Sélection aléatoire du film ---
    selected_movie_info = None
    while True:
        print("Voulez-vous choisir un film aléatoirement parmi la liste ? (oui/non)")
        choice_random = input().lower().strip()
        if choice_random == 'oui':
            if 'df' in globals() and not df.empty:
                selected_movie_info = df.sample(1).iloc[0]
                print(f"Un film aléatoire a été choisi.")
            else:
                print("La liste de films n'est pas disponible ou est vide pour le choix aléatoire.")
            break
        elif choice_random == 'non':
            print("Vous avez choisi de ne pas sélectionner de film aléatoire.")
            break
        else:
            print("Choix invalide. Veuillez répondre 'oui' ou 'non'.")

    if selected_movie_info is None:
        print("Aucun film n'a été sélectionné pour le jeu. Fin du jeu.")
        return None, None

    correct_title = selected_movie_info['title'].lower().strip()
    available_hints = difficulty_hint_sequence[difficulty]
    revealed_hints_count = 0
    attempts_left = max_attempts_dict[difficulty]

    def display_current_hints(movie_info, hints_to_show_keys, hint_providers_dict):
        print("\n--- Indices du film ---")
        for hint_key in hints_to_show_keys:
            if hint_key in hint_providers_dict:
                print(hint_providers_dict[hint_key](movie_info))
        print("-----------------------")

    # Affichage initial des indices
    initial_hints_keys = available_hints[:initial_hints_count[difficulty]]
    display_current_hints(selected_movie_info, initial_hints_keys, hint_providers)
    revealed_hints_count = initial_hints_count[difficulty]

    # --- Partie 3: Jeu de devinette interactif ---
    while attempts_left > 0:
        print(f"Il vous reste {attempts_left} essai(s).")
        guess = input("\nQuel est le titre du film (ou 'abandonner' pour révéler) ? ").lower().strip()

        if guess == 'abandonner':
            print(f"Dommage ! Le titre du film était '{selected_movie_info['title']}'.")
            break
        elif guess == correct_title:
            print(f"Félicitations ! Vous avez trouvé le film : '{selected_movie_info['title']}'.")
            break
        else:
            attempts_left -= 1
            if attempts_left == 0:
                print("Ce n'est pas le bon titre.")
                print(f"Perdu ! Vous n'avez plus d'essais. Le titre du film était '{selected_movie_info['title']}'.")
            else:
                print("Ce n'est pas le bon titre. Réessayez !")
                # Révéler un nouvel indice si disponible
                if revealed_hints_count < len(available_hints):
                    next_hint_key = available_hints[revealed_hints_count]
                    print(f"Voici un nouvel indice : {hint_providers[next_hint_key](selected_movie_info)}")
                    revealed_hints_count += 1
                else:
                    print("Désolé, il n'y a plus d'indices supplémentaires disponibles.")

    return difficulty, selected_movie_info['title']
```

```python
import PySimpleGUI as sg
import random

def play_gui(df):
    # Choix aléatoire d'un film
    film = df.sample(1).iloc[0]
    titre_correct = film["title"]

    # Indices simples (tu peux en ajouter)
    indices = [
        f"Année de sortie : {film.get('release_date', 'NA')}",
        f"Budget : {film.get('budget', 'NA')}",
        f"Première lettre du titre : {titre_correct[0]}"
    ]

    # Interface graphique
    layout = [
        [sg.Text("🎬 Devinez le film", font=("Helvetica", 16))],
        [sg.Text(indices[0], key="-INDICE-")],
        [sg.Input(key="-REPONSE-")],
        [sg.Button("Valider"), sg.Button("Quitter")],
        [sg.Text("", key="-RESULTAT-")]
    ]

    window = sg.Window("Jeu des 5000 films", layout)

    tentative = 0

    while True:
        event, values = window.read()

        if event in (sg.WINDOW_CLOSED, "Quitter"):
            break

        if event == "Valider":
            reponse = values["-REPONSE-"].strip().lower()
            tentative += 1

            if reponse == titre_correct.lower():
                window["-RESULTAT-"].update("🎉 Gagné !")
                break
            else:
                if tentative < len(indices):
                    window["-RESULTAT-"].update("❌ Mauvaise réponse. Nouvel indice affiché.")
                    window["-INDICE-"].update(indices[tentative])
                else:
                    window["-RESULTAT-"].update(
                        f"❌ Perdu. Le film était : {titre_correct}"
                    )
                    break

    window.close()
```
